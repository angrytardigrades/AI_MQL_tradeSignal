```python
import numpy as np
import pandas as pd
import sklearn as skt
```

```python
data = pd.read_csv("report_final.csv")
df = pd.DataFrame(data)
```

```python

data["ccpi"]=data["ccpi"].round(decimals=2)
data["ma"]=data["ma"].round(decimals=5)
data["findex"]=data["findex"].round(decimals=5)
data["rvi"]=data["rvi"].round(decimals=5)
data["rsi"]=data["rsi"].round(decimals=5)
data_r=data.drop(["datetime","tvol","findex","rvi","rsi","spr"],axis=1)
data_r=data_r.drop(data.tail(100).index,axis=0)


```

```python
data_r
```

```python
data_r.to_csv("data_r.csv")
```

```python
X = data_r.drop("signal",axis=1)
y= data_r["signal"]
X
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```
